```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/kaggle/input/zomato.csv')
df.head()
```

```python
df.shape
```

```python
df = df.drop(columns = ['url', 'address','phone','reviews_list', 'menu_item','dish_liked'])
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if value == '-' or value == 'NEW':
        # value = 'nan'
        return np.nan
    else :
        value = str(value).split('/')
        value = value[0]
        return float(value)
```

```python
df['rate'] = df['rate'].apply(handlerate)
```

```python
df['rate'].head()
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.rename(columns = {'approx_cost(for two people)':'cost2plates' , 'listed_in(type)':'type'},inplace=True)
df.head()
```

```python
df.drop(columns = 'listed_in(city)', inplace=True)
```

```python
df['cost2plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    value = value.replace(',','')
    return int(value)
```

```python
df['cost2plates'].apply(handlecomma)
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts()
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_types(value):
    if value in rest_types_lessthan1000:
        return 'others'
    else :
        return value
df['rest_type'] = df['rest_type'].apply(handle_rest_types)
```

```python
df['rest_type'].value_counts()
```

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts()
```

```python
location_lessthan300 =  location[location<300]
location_lessthan300.head()
```

```python
def handle_location(value):
    if value in location_lessthan300:
        return 'others'
    else :
        return value
df['location'] = df['location'].apply(handle_location)
```
